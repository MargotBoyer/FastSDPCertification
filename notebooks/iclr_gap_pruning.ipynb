```python

data_index_all_pruning = tab_with_all_pruning['data_index'].values
data_index_no_pruning = tab_without_pruning['data_index'].values    
print("Number of examples with full pruning:", len(set(data_index_all_pruning)))
print("Number of examples with no pruning:", len(set(data_index_no_pruning)))

print('Data indices in no pruning but not in full pruning:', set(data_index_no_pruning) - set(data_index_all_pruning))
for i in data_index_no_pruning:
    if i not in data_index_all_pruning:
        print("Missing index in full pruning results:", i)
    
for i in data_index_all_pruning:
    if i not in data_index_no_pruning:
        print("Missing index in full pruning results:", i)
```

```python
if "RLT_prop" in tab_with_all_pruning.columns:
    tab_with_all_pruning["RLT_prop"] = tab_with_all_pruning["RLT_prop"].fillna(0)
    RLT_prop = tab_with_all_pruning["RLT_prop"].unique()
    
    print("RLT_prop:", RLT_prop)
    #RLT_props = [0, 0.1, 0.3, 0.5]
    #tab = tab[tab["RLT_prop"].isin(RLT_props)].copy()
    

tab_with_all_pruning["total_time"] = tab_with_all_pruning["time"] + tab_with_all_pruning["bound_time"]
#tab_with_all_pruning["target"] = tab_with_all_pruning["target"].fillna("unknown")

if "RLT_prop" in tab_without_pruning.columns:
    tab_without_pruning["RLT_prop"] = tab_without_pruning["RLT_prop"].fillna(0)
    RLT_prop = tab_without_pruning["RLT_prop"].unique()
    
    print("RLT_prop:", RLT_prop)
    #RLT_props = [0, 0.1, 0.3, 0.5]
    #tab = tab[tab["RLT_prop"].isin(RLT_props)].copy()
    

tab_without_pruning["total_time"] = tab_without_pruning["time"] + tab_without_pruning["bound_time"]
tab_without_pruning["robust"] = (tab_without_pruning["optimal_value"] >= 0) | (tab_without_pruning["status"]=="trivially_solved")
```

```python
tab_with_all_pruning["optimal_value_partial_pruning"] = tab_with_all_pruning.apply(lambda row : tab_without_pruning[tab_without_pruning["data_index"]==row["data_index"]]["optimal_value"].values[0], axis=1)
```

```python
tab_with_all_pruning["gap_pruning"] = (tab_with_all_pruning["optimal_value"] - tab_with_all_pruning["optimal_value_partial_pruning"])
tab_with_all_pruning["rel_gap_pruning"] = tab_with_all_pruning["gap_pruning"] / np.abs(tab_with_all_pruning["optimal_value_partial_pruning"])
tab_with_all_pruning["robust_full_puning"] = (tab_with_all_pruning["optimal_value"] >= 0) 
tab_with_all_pruning["robust_partial_pruning"] = (tab_with_all_pruning["optimal_value_partial_pruning"] >= 0)
```
