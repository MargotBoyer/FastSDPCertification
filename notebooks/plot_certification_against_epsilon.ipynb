# Certification performances against pertubation bounds

```python

import matplotlib.pyplot as plt
import numpy as np

epsilon = np.array([0.026, 0.05, 0.1])
SDPt_time = np.array([399, 3977,  8890])
SDPu_fp_time = np.array([323, 1391, 3820])
beta_CROWN_time = np.array([2, 3, 3])
SDPt_verified = [74,6,0]
SDPu_fp_verified = [86,84,62]
beta_CROWN_verified = [69,3,0]
```

```python
from matplotlib.lines import Line2D

fig, ax1 = plt.subplots()

linestyles = ['-', '--', ':']

color = 'tab:red'
ax1.set_xlabel('Epsilon')
ax1.set_ylabel('Time (s)', color='red')
ax1.plot(epsilon, SDPt_time,  color='red', linewidth=2, linestyle = '--',  label='SDPt time')
ax1.plot(epsilon, SDPu_fp_time, color='red', linewidth=2,  linestyle = '-', label='SDPu time')
ax1.plot(epsilon, beta_CROWN_time,  color='red', linewidth=2,  linestyle = ':', label='beta-CROWN time')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second Axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Verified percentage', color=color)  # we already handled the x-label with ax1
ax2.plot(epsilon, SDPt_verified,  color='blue', linewidth=2, linestyle = '--',  label='SDPt')
ax2.plot(epsilon, SDPu_fp_verified,  color='blue', linewidth=2, linestyle = '-', label='SDPu')
ax2.plot(epsilon, beta_CROWN_verified, color='blue', linewidth=2,  linestyle = ':', label='beta-CROWN')
ax2.tick_params(axis='y', labelcolor=color)


#  Lgende manuelle pour les styles de ligne
custom_lines = [
    Line2D([0], [0], color='black', linestyle='-', lw=2),
    Line2D([0], [0], color='black', linestyle='--', lw=2),
    Line2D([0], [0], color='black', linestyle=':', lw=2),
]

plt.legend(custom_lines, ['SDPu', 'SDPt', 'beta-CROWN'],
           loc='upper center', bbox_to_anchor=(0.5, -0.15),
           ncol=3, frameon=False, title="Type de mthode")

ax1.grid(True, which='both', linestyle='--', alpha=0.4)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# Exemple de donnes
epsilon = np.linspace(0, 1, 10)

# 3 courbes pour le temps
time1 = np.random.rand(10) * 10
time2 = np.random.rand(10) * 8
time3 = np.random.rand(10) * 12

# 3 courbes pour le % de vrification
verif1 = np.random.rand(10) * 100
verif2 = np.random.rand(10) * 100
verif3 = np.random.rand(10) * 100

# Cration de la figure
fig, ax1 = plt.subplots(figsize=(8, 5))

# Axe pour les temps
ax1.set_xlabel(' (epsilon)')
ax1.set_ylabel('Temps de calcul (s)', color='tab:blue')
ax1.set_title('Temps de calcul et % de vrification en fonction de ')

# Diffrents styles de ligne
linestyles = ['-', '--', ':']

# Couleurs
colors_time = ['tab:blue', 'tab:orange', 'tab:green']
colors_verif = ['tab:red', 'tab:purple', 'tab:brown']

# Tracs
ax1.plot(epsilon, time1, linestyle=linestyles[0], marker='o', color=colors_time[0], label='Temps mthode 1')
ax1.plot(epsilon, time2, linestyle=linestyles[1], marker='s', color=colors_time[1], label='Temps mthode 2')
ax1.plot(epsilon, time3, linestyle=linestyles[2], marker='^', color=colors_time[2], label='Temps mthode 3')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Deuxime axe pour les % de vrification
ax2 = ax1.twinx()
ax2.set_ylabel('% de vrification', color='tab:red')

ax2.plot(epsilon, verif1, linestyle=linestyles[0], marker='o', color=colors_verif[0], label='% vrif 1')
ax2.plot(epsilon, verif2, linestyle=linestyles[1], marker='s', color=colors_verif[1], label='% vrif 2')
ax2.plot(epsilon, verif3, linestyle=linestyles[2], marker='^', color=colors_verif[2], label='% vrif 3')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Fusion des lgendes automatiques (pour les courbes)
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()

#  Lgende hors du graphique ( droite)
ax1.legend(lines1 + lines2, labels1 + labels2,
            loc='center left', bbox_to_anchor=(1.05, 0.5),
            frameon=False, title="Courbes")

#  Lgende manuelle pour les styles de ligne
custom_lines = [
    Line2D([0], [0], color='black', linestyle='-', lw=2),
    Line2D([0], [0], color='black', linestyle='--', lw=2),
    Line2D([0], [0], color='black', linestyle=':', lw=2),
]

plt.legend(custom_lines, ['Mthode 1', 'Mthode 2', 'Mthode 3'],
           loc='upper center', bbox_to_anchor=(0.5, -0.15),
           ncol=3, frameon=False, title="Type de mthode")

ax1.grid(True, which='both', linestyle='--', alpha=0.4)
plt.tight_layout()

plt.show()
```
