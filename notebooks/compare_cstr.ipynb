```python
def compare_presence(file1, file2):
    for constraint1 in file1:
        found = False
        for constraint2 in file2:
            if constraint1 == constraint2:
                #print(f"Constraint {constraint1.strip()} is the same in both files")
                found = True
                break
        if not found:
            print(f"Constraint {constraint1.strip()} is only in the first file")
```

```python
def count_duplicates(file_name):
    all_lines = []
    with open(file_name, 'r', encoding='utf-8') as file:
            for line in file:
                line = line.strip()
                if line:  # Skip empty lines
                    all_lines.append(line)

    total_lines = len(all_lines)
    unique_lines = len(set(all_lines))
    duplicates = total_lines - unique_lines
    
    return total_lines, unique_lines, duplicates
```
